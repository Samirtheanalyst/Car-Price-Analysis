```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\Samir\\OneDrive\\Desktop\\car_price_dataset.csv")
```

```python
# Question 1: Average Price by Brand
avg_price_by_brand = df.groupby('Brand')['Price'].mean().round(2).reset_index()
print("Average Price by Brand:\n", avg_price_by_brand)
```

```python
# Visualization 1: Bar chart for Average Price by Brand
plt.figure(figsize=(10, 6))
plt.bar(avg_price_by_brand['Brand'], avg_price_by_brand['Price'], color='grey')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.title('Average Price by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question 2: Most Common Fuel Type
fuel_type_distribution = df['Fuel_Type'].value_counts().reset_index()
fuel_type_distribution.columns = ['Fuel_Type', 'Total_Cars']
print("\nMost Common Fuel Type:\n", fuel_type_distribution)
```

```python
# Visualization 2: Pie chart for Most Common Fuel Type
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_distribution['Total_Cars'], labels=fuel_type_distribution['Fuel_Type'], autopct='%1.1f%%', startangle=140)
plt.title('Most Common Fuel Type')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Question 3: Top 5 Expensive Models
top_5_expensive = df.groupby(['Brand', 'Model'])['Price'].max().nlargest(5).reset_index()
print("\nTop 5 Expensive Models:\n", top_5_expensive)
```

```python
# Visualization 3: Horizontal bar chart for Top 5 Expensive Models
plt.figure(figsize=(10, 6))
plt.barh(top_5_expensive['Brand'] + ' ' + top_5_expensive['Model'], top_5_expensive['Price'], color='grey')
plt.xlabel('Price')
plt.ylabel('Brand and Model')
plt.title('Top 5 Expensive Models')
plt.tight_layout()
plt.show()
```

```python
# Question 4: Price vs. Mileage Correlation
df['Mileage_Category'] = df['Mileage'].apply(lambda x: 'High Mileage' if x > 100000 else 'Low Mileage')
price_vs_mileage = df.groupby('Mileage_Category')['Price'].mean().round(2).reset_index()
print("\nPrice vs. Mileage Correlation:\n", price_vs_mileage)
```

```python
# Visualization 4: Bar chart for Price vs. Mileage Correlation
plt.figure(figsize=(8, 6))
plt.bar(price_vs_mileage['Mileage_Category'], price_vs_mileage['Price'], color='grey')
plt.xlabel('Mileage Category')
plt.ylabel('Average Price')
plt.title('Price vs. Mileage Correlation')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question 5: New vs. Older Cars
df['Age_Category'] = df['Year'].apply(lambda x: 'New (After 2020)' if x > 2020 else 'Older (2020 or Earlier)')
new_vs_older = df.groupby('Age_Category')['Price'].mean().round(2).reset_index()
print("\nNew vs. Older Cars:\n", new_vs_older)
```

```python
# Visualization 5: Pie chart for New vs. Older Cars
plt.figure(figsize=(8, 8))
plt.pie(new_vs_older['Price'], labels=new_vs_older['Age_Category'], autopct='%1.1f%%', startangle=140)
plt.title('New vs. Older Cars')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Question 6: Transmission Preference
transmission_preference = df.groupby('Transmission').agg(
    Total_Cars=('Transmission', 'size'),
    Average_Price=('Price', 'mean')
).round(2).reset_index()
print("\nTransmission Preference:\n", transmission_preference)
```

```python
# Visualization 6: Stacked bar chart for Transmission Preference
plt.figure(figsize=(10, 6))
plt.bar(transmission_preference['Transmission'], transmission_preference['Total_Cars'], label='Total Cars', color='lightskyblue')
plt.bar(transmission_preference['Transmission'], transmission_preference['Average_Price'], label='Average Price', color='lightcoral')
plt.xlabel('Transmission')
plt.ylabel('Count/Average Price')
plt.title('Transmission Preference')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Question 7: Impact of Ownership on Price
ownership_impact = df.groupby('Owner_Count')['Price'].mean().round(2).reset_index()
print("\nImpact of Ownership on Price:\n", ownership_impact)
```

```python
# Visualization 7: Line chart for Impact of Ownership on Price
plt.figure(figsize=(8, 6))
plt.plot(ownership_impact['Owner_Count'], ownership_impact['Price'], marker='o', color='black', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Number of Owners')
plt.ylabel('Average Price')
plt.title('Impact of Ownership on Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Question 8: Doors vs. Price
doors_vs_price = df.groupby('Doors')['Price'].mean().round(2).reset_index()
print("\nDoors vs. Price:\n", doors_vs_price)
```

```python
# Visualization 8: Bar chart for Doors vs. Price
plt.figure(figsize=(8, 6))
plt.bar(doors_vs_price['Doors'].astype(str), doors_vs_price['Price'], color='grey')
plt.xlabel('Number of Doors')
plt.ylabel('Average Price')
plt.title('Doors vs. Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
